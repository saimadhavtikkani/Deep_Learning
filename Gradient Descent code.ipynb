```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python

def log_loss2(y,y_pred):
    epsilon = 1e-15
    y_pred_new = [max(i,epsilon) for i in y_pred]
    y_pred_new = [min(i,1-epsilon) for i in y_pred_new]
    y_pred_new = np.array(y_pred_new)
    return -np.mean(y*np.log(y_pred)+(1-y)*np.log(1-y_pred))

```

```python

def gradient_descent(x1,x2,y,epoch):
    w1 = w2 = 1
    b = 0
    rate = 0.5
    n = len(x1)
    for i in range(epoch):
        y_pred2 = w1*x1+w2*x2+b
        y_pred = sigmoid(y_pred2)
        loss = log_loss2(y,y_pred)
        
        w1d = (1/n)*np.dot(np.transpose(x1),(y_pred-y))
        w2d = (1/n)*np.dot(np.transpose(x2),(y_pred-y))
        bd =   np.mean(y_pred-y) 
        
        w1 = w1 - rate * w1d
        w2 = w2 - rate * w2d
        b = b - rate * bd
        print(i,w1,w2,b,loss)
    return w1,w2,b

```
